# Monte Carlo Simulation

Generate two uniform random numbers x and y between 0 and 1
Calculate x^2 + y^2
* If the value is less than 1 it's inside the circle
* If the value is more than 1 it's outside the circle
P(A) Favorable cases (A)/Possible cases
Use P(A) for aproximming PI value
Repit the the expirment enough times to get diferent approximations about PI
Calculate the average of the n numbers for giving a final value

```python
def piMontecarlo(n,m):    
    pi_avg = 0
    n =1000
    pi_value_list = []
    for i in range(m):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j]*x[j]+y[j]*y[j])
            if z<= 1:
                value += 1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value
    pi = pi_avg/m
    return pi

PI = piMontecarlo(10000,100)
print(PI)
```
