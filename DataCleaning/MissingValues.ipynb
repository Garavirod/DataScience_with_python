### Drop of missing values

```python
data.dropna(axis=0, how="all") #This erase all rows where nan is in all columns
```

```python
data.dropna(axis=0, how="any") #This erase all rows where nan is in any columns
```

### Implantation of missing values

```python
data.fillna("NO VALUE") #Implement an element in the missing values of the data frame
```

```python
data["dish_liked"].fillna("NO MENU")
```

### Median

```python
#Set in the missing values the mean of the exixtent data. only with numerical variables.
data["votes"].fillna(data["votes"].mean()) 
```

### backfill and ffill

```python
data["dish_liked"].fillna(method="ffill")
```

```python
data["dish_liked"].fillna(method="backfill")
```
