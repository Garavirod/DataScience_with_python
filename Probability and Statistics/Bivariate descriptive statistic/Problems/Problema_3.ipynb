#### Calculamos la media de X y Y

$ \bar{x} = \frac{\displaystyle\sum_{k=1}^N (x_{k})}{n} $


$ \bar{y} = \frac{\displaystyle\sum_{k=1}^N (y_{k})}{n} $

```python
x_median = ((20*20)+(45*50)+(75*30))/100
x_median
```

```python
y_median = ((10*35)+(30*65))/100
y_median
```

$ S_{x} = \sqrt{\frac{\displaystyle\sum_{k=1}^3 (n_{k}*x虏_{k})}{N} - \bar{x虏}}$ 

$ S_{y} = \sqrt{\frac{\displaystyle\sum_{k=1}^2 (n_{k}*y虏_{k})}{N} - \bar{y虏}}$ 

```python
sx = ceil(sqrt(((8000+101250+168750)/100)-(x_median**2)))
sx
```

```python
sy = ceil(sqrt((((3500+58500)/100)-y_median**2)))
sy
```
