# Boxplots (Box and whiskers)

Boxplots are a visual representation of data that describe several important characeristics, at the same time, such as __dispersion and symmetry__

```python
from IPython.display import Image
import numpy as np
from math import ceil
import matplotlib.pyplot as plt
```

#### Its representation is: 

Q1, Q2 (Median), Q3.


1. The first quartille Q1 and third Q3 form a box that represents an interval where it'll be 50% of the data.


2. Bellow and above median Q2(Quartille) there is 50% of data too.


3. The whiskers of the box are the min and max of the data.


4. Box's length is the __interquartile range__ ($RI$) and is defined like:
    
    
    
$ RI = Q3 - Q1 $

### Example

Given the scores of a set of 15 students on a math test

#### Seps:

1. Sort the data from low to highest.
2. Find Q1, Q2, and Q3
3. Find max and min value

```python
scores = [90,94,53,68,79,84,87,72,70,69,65,89,85,87,72]
```

```python
scores.sort()
print("Data sorted: ",scores)
Q1 = ceil(np.median(scores[0:len(scores)//2]))
Q2 = ceil(np.median(scores))
Q3 = ceil(np.median(scores[-len(scores)//2:]))

print("Quartille 1 Q1 :",Q1)
print("Quartille 2 Q2 (Median):",Q2)
print("Quartille 3 Q3 :",Q3)

print("Min :",min(scores))
print("Max :",max(scores))
```

### Using python tools

```python
groupA = [90,94,53,68,79,84,87,72,70,69,65,89,85,87,72]
groupB = [50,34,60,98,100,2,50,30,25,78,12,56,90,23,54]
#data array
data = [groupA,groupB]
 # Create a figure instance
fig = plt.figure(1, figsize=(15, 9))
# Create an axes instance
ax = fig.add_subplot(111)
#pint the box
bp = ax.boxplot(data, labels=["GROUP A","GROUP B"],patch_artist=True) # Create the boxplot
#Set title of box
plt.title("WHICH IS BETTER GROUP?")

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='lime', linewidth=3)
    # change fill color
    box.set( facecolor = 'maroon' )
        
# change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='navy', linewidth=2)
        
## change color and linewidth of the caps (max and min)
for cap in bp['caps']:
    cap.set(color='red', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='aqua', linewidth=3)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)    
```

### Atypical values

Values that are much bigger or much lower than the rest of data, that kind values are called *"outliers"*.
Outliers are represented with a point and can be in anywhere of diagram's edges.

To consider a value like atypical, value must be:

More than Q3, at least, 1.5 u of interquartile range.
Less than Q1, at least, 1.5 u of interquartile range.

That kind of limits are called barriers.

#### Superior and inferior barrier

Superior barrier:

$ BS = Q3 + 1.5 *RI $

Inferiror barrier:

$ BI = Q1 + 1.5 *RI $


### Symetry and how to calculate.

#### Symetry: 

When there is the same quantity of data in both sides of __arithmetic average__, median, arithmetic average and mode are equal.


#### Karl Pearson's Coefficient:


$ As = \frac{3(\bar{x}-Me)}{s} $

Karl Pearson's Coefficient varies between -3 and 3

1. $ As<0 $ Negative asymmetric (left).
2. $ As=0 $ Symetryc.
3. $ As>0 $ Postive asymetric (right).


#### Yule Bowley or Quartille Mesure:

$ As = \frac{Q_{1}+Q_{3}-2Q_{2}}{Q_{3}-Q_{1}} $


Yule Bowley or Quartille Mesure varies between -1 and 1

1. $ As<0 $ Negative asymmetric (left).
2. $ As=0 $ Symetryc.
3. $ As>0 $ Postive asymetric (right).


#### Fisher Mesure:

1. No grouped data:

    $ As = \frac{ \displaystyle\sum_{k=1}^N (x_{i}-\bar{x})^3}{ns^3} $
    
2. Grouped data with frequency:

    $ As = \frac{ \displaystyle\sum_{i=1}^k n_{i}*(x_{i}-\bar{x})^3}{ns^3} $        

3. Gruped data by intervals:

     $ As = \frac{ \displaystyle\sum_{i=1}^k n_{i}*(x_{mi}-\bar{x})^3}{ns^3} $

*Where $ n_{i} $ is the frequency*

*Where $ x_{mi} $ is the brand class*

*Where $ k $ is the number of class*
